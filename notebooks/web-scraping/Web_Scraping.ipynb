```python
# Ejemplo de Web Scraping en Google Colab usando Python

# En este ejemplo, vamos a realizar scraping del sitio web "Books to Scrape" (http://books.toscrape.com/), un sitio web especialmente diseñado para que los desarrolladores puedan practicar sus habilidades de scraping. Utilizaremos las bibliotecas `requests` y `BeautifulSoup` para acceder y analizar la información contenida en una página específica. A continuación, se explica cada línea de código a detalle.

# Importar las bibliotecas necesarias
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Paso 1: Hacer una solicitud HTTP a la URL deseada
# Definimos la URL que deseamos "scrapear"
url = "http://books.toscrape.com/"

# Utilizamos la biblioteca requests para hacer una solicitud GET al servidor y obtener el contenido de la página web
response = requests.get(url)

# Comprobamos que la solicitud haya sido exitosa verificando el estado de respuesta (200 significa OK)
if response.status_code == 200:
    print("Solicitud exitosa!")
else:
    print(f"Error al acceder a la URL. Estado de la solicitud: {response.status_code}")

# Paso 2: Parsear el HTML de la página con BeautifulSoup
# Ahora, utilizamos BeautifulSoup para analizar el contenido HTML de la página web
soup = BeautifulSoup(response.content, "html.parser")

# `soup` es un objeto que permite navegar y buscar elementos del HTML como si fuera un árbol

# Paso 3: Extraer información de interés
# En este caso, queremos extraer información de los libros, como título, autor, precio y disponibilidad

# Creamos listas vacías para almacenar la información extraída
book_titles = []
book_authors = []
book_prices = []
book_availabilities = []

# Recorremos todos los elementos que tienen la clase "product_pod" (cada libro)
for book in soup.find_all("article", class_="product_pod"):
    # Extraer el título del libro que está en la etiqueta "h3"
    title = book.h3.a["title"]
    book_titles.append(title)

    # Extraer el precio del libro que está en un elemento "p" con clase "price_color"
    price = book.find("p", class_="price_color").text
    book_prices.append(price)

    # Extraer la disponibilidad del libro que está en un elemento "p" con clase "instock availability"
    availability = book.find("p", class_="instock availability").text.strip()
    book_availabilities.append(availability)

# Paso 4: Mostrar los datos extraídos
# Convertimos las listas extraídas en un DataFrame para poder trabajar más cómodamente con los datos
books_df = pd.DataFrame({
    "Title": book_titles,
    "Price": book_prices,
    "Availability": book_availabilities
})

# Mostrar los primeros registros extraídos
books_df.head()

# Explicación detallada de cada línea de código
# 1. Importamos las bibliotecas necesarias: `requests` para hacer solicitudes HTTP, `BeautifulSoup` para analizar el contenido HTML, y `pandas` para organizar los datos en formato tabular.
# 2. Definimos la URL del sitio web que queremos "scrapear".
# 3. Realizamos una solicitud GET a la URL con `requests.get(url)`. Esto descarga el HTML completo de la página.
# 4. Verificamos si la solicitud fue exitosa comprobando el código de estado HTTP.
# 5. Parseamos el contenido HTML utilizando BeautifulSoup, que nos permite analizar y extraer información de manera estructurada.
# 6. Creamos listas vacías para almacenar los títulos, precios y disponibilidad de cada libro.
# 7. Utilizamos un bucle para recorrer cada artículo que contiene un libro (usando la clase "product_pod"), luego extraemos la información relevante como el título, precio y disponibilidad, y la almacenamos en las listas respectivas.
# 8. Creamos un DataFrame utilizando `pandas` para mostrar los datos de manera tabular.
# 9. Finalmente, mostramos los primeros registros extraídos del DataFrame para verificar los resultados.

# Nota: Este ejemplo puede ser ejecutado en Google Colab, asegurándose de tener las bibliotecas necesarias instaladas.
# Para instalar BeautifulSoup y requests en Google Colab, se pueden ejecutar los siguientes comandos:
# !pip install requests
# !pip install beautifulsoup4

# El web scraping debe realizarse siempre de manera ética y respetando los términos y condiciones del sitio web. Este ejemplo utiliza un sitio web diseñado específicamente para la práctica del scraping.
```
