
# 🧪 A/B Testing en Python — Script Completo para Google Colab

Este notebook implementa un experimento **A/B clásico** para evaluar si dos versiones (A y B) presentan diferencias estadísticamente significativas en su tasa de conversión.

Incluye:
- Simulación de datos  
- Visualización  
- Prueba estadística (Z-test)  
- Intervalo de confianza  
- Interpretación final  


```

# ============================================================
# 1. IMPORTACIÓN DE LIBRERÍAS
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

plt.style.use('seaborn-v0_8')
np.random.seed(42)
```

```

# ============================================================
# 2. SIMULACIÓN DEL EXPERIMENTO A/B
# ============================================================

n_A = 10000  # versión A
n_B = 10000  # versión B

p_A = 0.10
p_B = 0.12

conversions_A = np.random.binomial(1, p_A, n_A)
conversions_B = np.random.binomial(1, p_B, n_B)

rate_A = conversions_A.mean()
rate_B = conversions_B.mean()

print(f"Tasa de conversión A: {rate_A:.3f}")
print(f"Tasa de conversión B: {rate_B:.33f}")
print(f"Diferencia observada: {rate_B - rate_A:.3f}")
```

```

# ============================================================
# 3. VISUALIZACIÓN DE RESULTADOS
# ============================================================

plt.bar(['A', 'B'], [rate_A, rate_B], color=['skyblue', 'salmon'])
plt.title('Comparación de Tasas de Conversión')
plt.ylabel('Tasa de conversión')
plt.show()
```

```

# ============================================================
# 4. PRUEBA ESTADÍSTICA: DIFERENCIA DE PROPORCIONES
# ============================================================

success_A = conversions_A.sum()
success_B = conversions_B.sum()

p_pool = (success_A + success_B) / (n_A + n_B)
se = np.sqrt(p_pool * (1 - p_pool) * (1/n_A + 1/n_B))

z_score = (rate_B - rate_A) / se
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Estadístico Z: {z_score:.3f}")
print(f"p-valor: {p_value:.4f}")
```

```

# ============================================================
# 5. INTERPRETACIÓN DEL TEST
# ============================================================

alpha = 0.05

if p_value < alpha:
    print("✅ La diferencia es estadísticamente significativa. La versión B supera a la A.")
else:
    print("❌ No hay evidencia suficiente: no se detecta diferencia significativa.")
```

```

# ============================================================
# 6. INTERVALO DE CONFIANZA DEL 95%
# ============================================================

diff = rate_B - rate_A
z_crit = stats.norm.ppf(1 - alpha/2)

ci_low = diff - z_crit * se
ci_high = diff + z_crit * se

print(f"Intervalo de confianza (95%): [{ci_low:.4f}, {ci_high:.4f}]")
```


# 📘 Conclusión

Este experimento A/B demuestra cómo evaluar si dos versiones (A y B) presentan diferencias
estadísticamente significativas en su tasa de conversión.

La prueba Z permite verificar si la diferencia observada se debe al azar o a una mejora real.
El intervalo de confianza ayuda a cuantificar la magnitud posible de la mejora.
