# Procesamiento de datos

```python
# unir vars e income en un solo dataframe llamado df usando la función pd.concat()
df = pd.concat([vars,income],axis = 1)
```

```python
# obtener atributos del dataframe df, imprimirlos
# shape
print("Tamaño del dataframe:\n", df.shape)

# tipos de datos
print("\nTipo de datos:\n",df.dtypes)

# nombre de columnas
print("\nNombre de columnas: \n",df.columns)
```

```python
#checar si hay valores nulos usando las funciones isna() y sum() en ese orden
df.isna().sum()
```

```python
#si hay valores nulos, eliminarlos, y guardar en un nuevo dataframe df_2 usando dropna()
df_2 = df.dropna()

#revisar que sí se hayan eliminado los valores nulos
df_2.isna().sum()
```

```python
# checar valores duplicados utlizando la función duplicated(keep = False) en df_2 y guárdalo en una nueva variable llamado duplicados
# este va a generar una serie de True y False para luego poder identificar cuáles son las filas que tienen valores duplicados
duplicados = df_2.duplicated(keep = False)

# filtra df_2 por los valores duplicados y guárdalo en un nuevo dataframe valores_duplicados
valores_duplicados = df_2[duplicados]

# enseña valores_duplicados, ordena por las columnas age y fnlwgt usando la función sort_values y el parámetro by
valores_duplicados.sort_values(by = ["age","fnlwgt"])
```

```python
# elimina los duplicados del df_2 usando la función drop_duplicates() y guárdalo en un nuevo dataframe df_3
df_3 = df_2.drop_duplicates()

# imprime el shape de df_2 y df_3 en ese orden para ver que se hayan eliminado correctamente los valores duplicados
print(df_2.shape, df_3.shape)
```

```python
# Agrupar las columnas por tipo de datos
agrupar_por_dtype = df_3.columns.to_series().groupby(df_3.dtypes).groups

# Imprime los valores únicos de cada columna junto con su tipo de dato
for dtype, columns in agrupar_por_dtype.items():
    print(f'Tipo de dato: {dtype}')
    for col in columns:
        unique_values = df_3[col].unique()
        print(f'Columna: {col}\nValores únicos: {unique_values}\n')
```

Al revisar los valores únicos de las columnas, a primera vista nos damos cuenta que en las columnas `workclass`, `occupation` y `native-country` contienen un valor de "?"

También, podemos ver que en la columna de `income` tenemos categorías duplicadas '<=50K'/ con '<=50K.' y '>50K' con '>50K.'

Trabajemos para arreglar estas columnas.

```python
df_4 = df_3.copy()
# Reemplazar "?" con "Indefinido" en las columnas que identificamos usando la función replace
#define una lista de las columans que vamos a ajustar
columnas_a_ajustar = ["workclass", "occupation", "native-country"]
#utiliza la función replace para ajustar esas columnas
df_4.loc[:, columnas_a_ajustar] = df_4[columnas_a_ajustar].replace('?', 'Indefinido')

# Reemplazar valores en la columna income, <=50K. con <=50K y >50K. con >50K
df_4.loc[:, "income"] = df_4['income'].replace({'<=50K.': '<=50K', '>50K.': '>50K'})

# Imprimir  valores de la columna income para revisar que sí se hayan aplicado los cambios
unicos_income = df_4['income'].unique()
unicos_country = df_4['native-country'].unique()
print(f'Columna: income\nValores: {unicos_income}\n')
print(f'Columna: native-country\nValores: {unicos_country}\n')

df_4[df_4["native-country"] == "Indefinido"]
```

# Visualizaciones
Para esta última sección vamos a tener que realizar las siguientes visualizaciones:


**Gráfica de Barras:** Número de registros por nivel de educación

**Boxplot:** Edad por estado civil e ingreso

**Kernel Density Plot:** Distribución de edad por estado civil

```python
# Contar el número de filas para cada categoría de educación usando value_counts()
education_counts = df_usa_2['education'].value_counts()

plt.figure(figsize=(10, 6))  # Establecer el tamaño de la figura
# Crear un gráfico de barras utilizando plt.bar
plt.bar(education_counts.index, education_counts.values)

# Establecer etiquetas y título
plt.xlabel('Educación')
plt.ylabel('Número de Registros')
# Define el título como 'Número de Registros por nivel de educación'
plt.title("Número de Registros por nivel de educación")

# Rotar etiquetas del eje x  por 45 grados para una mejor visibilidad
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.tight_layout()  # Ajustar el diseño para una mejor apariencia
plt.show()
```

📊📈 ANÁLISIS 📊📈

**Describe una conclusión que puedas derivar de la gráfica anterior:**

*RESPUESTA*

1. Educación Secundaria como Mayoría: El nivel con el mayor número de registros es la educación secundaria (High school), lo que sugiere que la mayoría de las personas en este conjunto de datos han completado al menos la educación secundaria.

2. Disminución con Niveles de Educación Avanzados: A medida que aumenta el nivel de educación, el número de registros disminuye. Esto es evidente en los niveles de educación postsecundaria, donde los que tienen títulos de bachillerato son menos que los de educación secundaria, y sigue disminuyendo para los niveles de maestría y doctorado.

3. Educación Primaria y Preescolar Menos Comunes: Los niveles de educación primaria y preescolar tienen muy pocos registros en comparación con los niveles de educación secundaria y postsecundaria. Esto podría reflejar un énfasis en la base de datos hacia la población adulta o una menor incidencia de individuos con niveles de educación bajos.

4. Educación Profesional y Vocacional: Los títulos profesionales y vocacionales (Prof-school, Assoc-voc) tienen una presencia modesta en comparación con la educación universitaria. Esto podría indicar que hay una tendencia menor hacia la educación especializada o que el conjunto de datos no captura completamente a la población con este tipo de educación.

```python
sns.set_theme(style="ticks", palette="bright")

# Crear un boxplot utilizando Seaborn función boxplot
# - El eje x es "age"
# - El eje y es "marital-status"
# - Hue (color) está determinado por "income"
# - Los datos provienen del DataFrame df_usa_2
sns.boxplot(x="age", y="marital-status",
            hue ="income", palette=["m", "g"],
            data=df_usa_2)

sns.despine(offset=10, trim=True)
```

📊📈 ANÁLISIS 📊📈

**Describe una conclusión que puedas derivar de la gráfica anterior:**

*RESPUESTA*

1. Ingresos en Relación al Estado Civil: Para la mayoría de los estados civiles, las personas con ingresos superiores a 50,000 parecen ser mayores que aquellas con ingresos inferiores a 50,000. Esto podría sugerir que mayores ingresos están asociados con una mayor edad en varios estados civiles.

2. Nunca Casados: En el grupo de nunca casados, la mediana de edad para ambos segmentos de ingresos es relativamente joven, y la distribución de edad es más estrecha para aquellos con ingresos superiores a 50,000, lo que podría indicar que un mayor ingreso en este grupo está asociado con un rango de edad más concentrado.

3. Divorciados y Separados: Los divorciados y separados presentan una amplia distribución de edad, pero con una mediana que tiende a estar en la mitad de la vida adulta. Esto coincide con la noción de que el divorcio y la separación suelen ocurrir después de varios años de matrimonio.

4. Viudos y Casados con Cónyuge Ausente: La viudez y estar casado con cónyuge ausente muestran una mediana de edad avanzada, lo cual es esperable dado que estos estados civiles son más comunes a medida que las personas envejecen. Además, para los viudos, no parece haber una gran diferencia en la distribución de edad entre los dos segmentos de ingreso.

```python
# Configuración de Seaborn para un fondo blanco y una figura sin color de fondo
sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

# Definir una paleta personalizada con tonos de naranja excluyendo amarillo
custom_palette = sns.color_palette("YlOrRd", n_colors=5)[1:]

# Aumentar el tamaño de la figura para dar más espacio a los subplots
plt.figure(figsize=(16, 9))

# Inicializar el objeto FacetGrid row y hue marital-status
g = sns.FacetGrid(df_usa_2, row="marital-status", hue="marital-status", aspect=15, height=1, palette=custom_palette)

# Dibujar los kernel density plots usando la columna age
g.map(sns.kdeplot, "age",
      bw_adjust=.5, clip_on=False,
      fill=True, alpha=1, linewidth=1.5)

g.map(sns.kdeplot, "age", clip_on=False, color="w", lw=2, bw_adjust=.5)

# Pasar color=None a refline() utiliza la asignación de color 'hue'
g.refline(y=0, linewidth=2, linestyle="-", color=None, clip_on=False)

# Definir y usar una función simple para etiquetar el gráfico en coordenadas del eje
def label(x, color, label):
    ax = plt.gca()
    ax.text(1, .2, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes)

g.map(label, "age")

# Ajustar las subtramas para que se superpongan
g.figure.subplots_adjust(hspace=-.25)

# Eliminar detalles de los ejes que no se integran bien con la superposición
g.set_titles("")
g.set(yticks=[], ylabel="")
g.despine(bottom=True, left=True)

# Mostrar el gráfico
plt.show()
```

📊📈 ANÁLISIS 📊📈

**Describe una conclusión que puedas derivar de la gráfica anterior:**

*RESPUESTA*

1. Pico de Nunca Casados: Hay un pico prominente de individuos que nunca se han casado en el rango de edad de 20 a 30 años. Esto sugiere que una gran proporción de jóvenes adultos están solteros.

2. Casados con Cónyuge Civil: La categoría de casados con cónyuge civil parece ser la más común en todas las edades después de los 30 años, indicando que el matrimonio es el estado civil predominante en la adultez media.

3. Divorcios y Separaciones: Hay un aumento en la proporción de personas divorciadas y separadas que comienza alrededor de los 30 años y parece tener un pico alrededor de los 50 años. Esto podría reflejar que los divorcios son más comunes a medida que las personas llegan a la mediana edad.

4. Viudez y Casados Ausentes: La viudez aumenta significativamente con la edad, lo cual es esperado debido a la pérdida de cónyuges a edades más avanzadas. La categoría de casados con cónyuge ausente también está presente pero es menos común y no muestra un patrón de edad tan claro como las otras categorías.
