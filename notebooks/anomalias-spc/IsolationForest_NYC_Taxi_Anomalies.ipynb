```
# DETECCIÓN DE ANOMALÍAS CON ISOLATION FOREST
# CASO DE USO: DEMANDA DE TAXIS EN NYC
```

```
# CASO DE USO:
# Este script analiza los datos históricos de la demanda de taxis en la ciudad de Nueva York
# para detectar comportamientos anómalos, como caídas o picos abruptos. Estas anomalías pueden deberse a eventos extraordinarios
# como tormentas de nieve, celebraciones masivas (Año Nuevo), apagones, etc.
# El objetivo es desarrollar una herramienta basada en aprendizaje automático no supervisado que identifique automáticamente estos puntos atípicos.
```

```
# 1. IMPORTACIÓN DE LIBRERÍAS
```

```
import matplotlib.pyplot as plt           # Para visualizaciones estáticas
import numpy as np                        # Para cálculos numéricos
import pandas as pd                       # Para manipulación de datos
from sklearn.ensemble import IsolationForest  # Algoritmo de detección de anomalías

import plotly.express as px               # Para visualizaciones interactivas

# Configuramos el tamaño por defecto de las gráficas de matplotlib
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['axes.grid'] = True
```

```
# 2. CARGA DEL CONJUNTO DE DATOS
```

```
# Cargamos los datos desde el repositorio público de NAB (Numenta Anomaly Benchmark)
url = 'https://raw.githubusercontent.com/numenta/NAB/master/data/realKnownCause/nyc_taxi.csv'
df = pd.read_csv(url)

# Mostramos una muestra de los datos para inspección inicial
df.head()
```

```
# 3. PREPROCESAMIENTO DE DATOS
```

```
# Convertimos la columna de timestamp (fecha y hora) a objeto datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Verificamos la estructura del dataframe
df.info()
```

```
# 4. ENTRENAMIENTO DEL MODELO ISOLATION FOREST
```

```
# Creamos el modelo. El parámetro 'contamination' indica el porcentaje estimado de anomalías en el dataset
model = IsolationForest(contamination=0.01, random_state=42)

# Ajustamos el modelo usando únicamente la variable 'value' (número de pasajeros por hora)
model.fit(df[['value']])

# Obtenemos las predicciones: -1 indica anomalía, 1 indica valor normal
df['anomaly'] = model.predict(df[['value']])
```

```
# 5. VISUALIZACIÓN DE ANOMALÍAS DETECTADAS
```

```
# Graficamos la serie de tiempo completa con Plotly para visualización interactiva
fig = px.line(df, x='timestamp', y='value', title='Demanda de taxis en NYC con detección de anomalías')

# Agregamos los puntos anómalos a la gráfica como marcadores
fig.add_scatter(
    x=df['timestamp'][df['anomaly'] == -1],
    y=df['value'][df['anomaly'] == -1],
    mode='markers',
    name='Anomalías',
    marker=dict(color='red', size=8, symbol='x')
)

# Mostramos la figura
fig.show()
```
